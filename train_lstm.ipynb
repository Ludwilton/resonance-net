```python
from src.rnet_model import AmpModel
from src.train import train_model
from src.utils import process_audio, evaluate_model
from pathlib import Path
import torch
import datetime

```

```python
input_file = "data/input.wav"
output_file = "data/output.wav"
```

```python
# train model
model = train_model(input_file, output_file, model_save_path="models/amp_model.pth")
```

```python
device = torch.device('cuda')

checkpoint = torch.load('models/amp_model.pth', map_location=device, weights_only=False)
hidden_size = checkpoint.get('hidden_size', 32)
model = AmpModel(hidden_size=hidden_size)
model.load_state_dict(checkpoint['model_state_dict'])
model.to(device)
model.eval()
```

```python
time_stamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
process_audio(
    model=model, 
    input_file="data/input.wav", 
    output_file=f"data/model_outputs/model_output_{time_stamp}.wav", 
    buffer_size=32, 
    device=device
)
```

```python

data_dir = Path("data")
models_dir = Path("models")
TARGET_WAV = data_dir / "output.wav"
PRED_WAV   = data_dir / "model_output_v2.wav"
evaluate_model(TARGET_WAV, PRED_WAV, snippet_start_sec=2)
```
